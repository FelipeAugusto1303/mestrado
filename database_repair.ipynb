```python
import glob
import os
import cv2 as cv

BASE_PATH='./newDataset'
CLASS_LEN=2200
```

```python
#
classes = []
folders = []
for image in glob.glob('./dataset/train_pose/*'):
    i = i + 1
    print(image.split("\\")[-1][3:])
    folders.append(image.split("\\")[-1])
    classes.append(image.split("\\")[-1][3:])
    
print(i)
print(classes)
print(folders)

for image in glob.glob('./dataset/train_pose/'+folders[0]+'/*'):
    print(image)
```

```python
# utils
def create_database_folder(folder_name):
    if not os.path.exists(os.path.join(BASE_PATH,folder_name)):
        os.mkdir(os.path.join(BASE_PATH,folder_name))
    
```

```python
#Making a training database
for i in range(0,16):
    create_database_folder(classes[i]) # create all folders database
    j = 0
    for image in glob.glob('./dataset/train_pose/'+folders[i]+'/*'):
        print(image)
        
        im = cv.imread(image)
        cv.imwrite(os.path.join(BASE_PATH,classes[i],classes[i]+'_'+str(j)+'.png'), im)
        j= j +1
        
    for image in glob.glob('./dataset/test_pose/'+folders[i]+'/*'):
        print(image)
        im = cv.imread(image)
        cv.imwrite(os.path.join(BASE_PATH,classes[i],classes[i]+'_'+str(j)+'.png'), im)
        j= j +1
        
```

```python
#Making a rebalance in database
for i in range(0,16):
    j = 0
    for image in glob.glob(BASE_PATH+'/'+classes[i]+'/*.png'):
        if j >= CLASS_LEN:
            os.remove(image)
        j = j + 1
            
```
